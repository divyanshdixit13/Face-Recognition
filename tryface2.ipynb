```python
import tkinter as tk
import cv2
from PIL import Image, ImageTk
import os
import subprocess

class App:
    def __init__(self):
        # Add the directory containing the Face Recognition executable to the PATH variable
        os.environ['PATH'] += os.pathsep + 'C:\\Users\\5CD1478RQY!\\Documents\\db'

        self.main_window = tk.Tk()
        self.main_window.geometry("1200x520+350+100")
        self.create_button()
        self.create_webcam_label()
        self.cap = cv2.VideoCapture(0)
        self.process_webcam()

        self.db_dir = './db'
        if not os.path.exists(self.db_dir):
            os.mkdir(self.db_dir)

    def start(self):
        self.main_window.mainloop()

    def login(self):
        unknown_img_path = './.tmp.jpg'
        cv2.imwrite(unknown_img_path,self.frame)
        output = subprocess.check_output(['face_recognition', self.db_dir,unknown_img_path])
        print(output)
        os.remove(unknown_img_path)

    def Register_New_User(self):
        self.register_new_user_window = tk.Toplevel(self.main_window)
        self.register_new_user_window.geometry("800x600+370+120")
        text_label_text = "Please input username:"
        self.text_label_register_new_user = tk.Label(self.register_new_user_window, text=text_label_text)
        self.text_label_register_new_user.place(x=100, y=560)
        self.entry_text_register_new_user = tk.Entry(self.register_new_user_window)
        self.entry_text_register_new_user.place(x=280, y=560)
        self.capture_label()

    def create_button(self):
        button = tk.Button(self.main_window, text="Login", bg='green', command=self.login)
        button.place(x=750, y=300)
        button1 = tk.Button(self.main_window, text="Register New User", bg='gray', command=self.Register_New_User, fg='black')
        button1.place(x=750, y=350)

    def create_webcam_label(self):
        self.webcam_label = tk.Label(self.main_window)
        self.webcam_label.place(x=10, y=0, width=700, height=500)

    # def process_webcam(self):
    #     ret, frame = self.cap.read()
    #     if ret:
    #         img_ = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    #         img = Image.fromarray(img_)

    #         # Create a new ImageTk.PhotoImage object
    #         imgtk = ImageTk.PhotoImage(image=img)

    #         # Keep a reference to the image object
    #         self.imgtk = imgtk

    #         # Configure the label with the new image
    #         self.webcam_label.configure(image=self.imgtk)

    #     self.main_window.after(10, self.process_webcam)

    def login(self):
        try:
            unknown_img_path = './.tmp.jpg'
            ret, frame = self.cap.read()
            if ret:
                cv2.imwrite(unknown_img_path, frame)
                output = subprocess.check_output(['face_recognition', self.db_dir, unknown_img_path])
                print(output)
                os.remove(unknown_img_path)
        except Exception as e:
            print("Error during login:", e)

    
    def process_webcam(self):
        ret, frame = self.cap.read()
        if ret:
            img_ = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            img = Image.fromarray(img_)
            self.imgtk = ImageTk.PhotoImage(image=img)  # Keep a reference to the ImageTk object
            self.webcam_label.configure(image=self.imgtk)
        self.main_window.after(10, self.process_webcam)

    def capture_label(self):
        capture_label = tk.Label(self.register_new_user_window)
        capture_label.place(x=50, y=50, width=700, height=500)
        self.add_img_to_label(capture_label)
        accept_button = tk.Button(self.register_new_user_window, text="Accept", bg='green', command=self.accept_register_new_user)
        accept_button.place(x=350, y=560)
        try_again_button = tk.Button(self.register_new_user_window, text="Try Again", bg='red', command=self.try_again_register_new_user)
        try_again_button.place(x=450, y=560)

    def add_img_to_label(self, label):
        ret, frame = self.cap.read()
        if ret:
            img_ = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            img = Image.fromarray(img_)
            self.imgtk = ImageTk.PhotoImage(image=img)  # Keep a reference to the ImageTk object
            label.imgtk = self.imgtk
            label.configure(image=self.imgtk)
        self.main_window.after(10, lambda: self.add_img_to_label(label))

    def accept_register_new_user(self):
        name = self.entry_text_register_new_user.get()
        ret, frame = self.cap.read()
        if ret:
            cv2.imwrite(os.path.join(self.db_dir, '{}.jpg'.format(name)), frame)
        print('Success', 'User Registered Successfully')
        self.register_new_user_window.destroy()

    def try_again_register_new_user(self):
        self.register_new_user_window.destroy()

if __name__ == "__main__":
    app = App()
    app.start()
```
